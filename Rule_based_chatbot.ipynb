```python
#Python code to implement rule-based chatbot

import random
import re

class AliensBot:
  neg_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")

  exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")

  #Random questions asked by the bot
  rand_questions = [
      "Why are you here? ",
      "Are there many humans like you? ",
      "What do you consume for sustenance? ",
      "Is there intelligent life on this planet? ",
      "Does Earth have a leader? ",
      "What planets have you visited? ",
      "What technology do you have on this planet? "
  ]

  def __init__(self):
    self.alienbabble = {
        #regular expression
        'describe_planet_intent' : r'.*\s*your\splanet',
        'answer_why_intent' : r'why\sare.*',
        'squared_intent' : r'.*square.*\s(\d+)'
    }

  def greet(self):
    self.name = input('what is your name?\n')
    will_help = input(f'Hi {self.name}, I am not from this planet. Could you please help me know about your planet?\n').lower()

    # If the response is negative conversation ends
    if will_help in self.neg_responses:
      print("It's ok, have a nice day!!")
      return

    #If the response is positive then chat continues
    self.chat()

  def make_exit(self,reply):
    if reply in self.exit_commands:
      print('Ok, have a nice day!!')
      return True

    return False

  def chat(self):
    #random question is picked from the list
    reply = input(random.choice(self.rand_questions))

    #If reply is positive then chat continues
    while not self.make_exit(reply):
      reply = input(self.match_reply(reply.lower()))

  def match_reply(self,reply):
    for key, value in self.alienbabble.items():
      entity = value
      intent = key

      #Command to match regular expressions
      will_match = re.match(entity, reply)

      if will_match and intent=='answer_why_intent':
        return self.answer_why_intent()
      elif will_match and intent=='describe_planet_intent':
        return self.describe_planet_intent()
      elif will_match and intent=='squared_intent':
        return self.squared_intent(will_match.group(1))

    if not will_match:
      return self.no_match_intent()

  def answer_why_intent(self):
    responses = ("I come in peace\n", "I am here to collect data on your planet and its inhabitants\n",
                     "I heard that pizza is good here\n")

    #selectind random responses
    return random.choice(responses)

  def describe_planet_intent(self):
    responses = ("My planet is a utopia of diverse organisms and species.\n",
                     "I am from Opidipus, the capital of the Wayward Galaxies.\n")

    return random.choice(responses)

  def squared_intent(self, number):
    num = int(number)
    return f'The square of the number {num} is {num*num}.\n'

  def no_match_intent(self):
    responses = (
        "Please tell me more.\n", "Tell me more!\n", "Why do you say that?\n", "I see. Can you elaborate?\n",
        "Interesting. Can you tell me more?\n", "I see. How do you think?\n", "Why?\n",
        "How do you think I feel when you say that?\n")

    return random.choice(responses)

alien = AliensBot()
alien.greet()
```
